```python
import numpy as np

perch_length = np.array(
    [8.4, 13.7, 15.0, 16.2, 17.4, 18.0, 18.7, 19.0, 19.6, 20.0,
     21.0, 21.0, 21.0, 21.3, 22.0, 22.0, 22.0, 22.0, 22.0, 22.5,
     22.5, 22.7, 23.0, 23.5, 24.0, 24.0, 24.6, 25.0, 25.6, 26.5,
     27.3, 27.5, 27.5, 27.5, 28.0, 28.7, 30.0, 32.8, 34.5, 35.0,
     36.5, 36.0, 37.0, 37.0, 39.0, 39.0, 39.0, 40.0, 40.0, 40.0,
     40.0, 42.0, 43.0, 43.0, 43.5, 44.0]
     )
perch_weight = np.array(
    [5.9, 32.0, 40.0, 51.5, 70.0, 100.0, 78.0, 80.0, 85.0, 85.0,
     110.0, 115.0, 125.0, 130.0, 120.0, 120.0, 130.0, 135.0, 110.0,
     130.0, 150.0, 145.0, 150.0, 170.0, 225.0, 145.0, 188.0, 180.0,
     197.0, 218.0, 300.0, 260.0, 265.0, 250.0, 250.0, 300.0, 320.0,
     514.0, 556.0, 840.0, 685.0, 700.0, 700.0, 690.0, 900.0, 650.0,
     820.0, 850.0, 900.0, 1015.0, 820.0, 1100.0, 1000.0, 1100.0,
     1000.0, 1000.0]
     )
```

```python
import matplotlib.pyplot as plt
plt.scatter(perch_length, perch_weight)
plt.xlabel("length")
plt.ylabel("weight")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(perch_length, perch_weight)
```

```python
# 1차원 넘파이 배열은 학습 불가 -> 2차원 넘파이 배열로 변환이 필요!
test_array = np.array([10, 20, 30, 40])
print(test_array)
test_array = test_array.reshape(2, 2)  # 2행 2열 변환
print(test_array)
# test_array = test_array.reshape(4, 1) # 4행 1열 변환
test_array = test_array.reshape(-1, 1) # -1은 각 요소 갯수 전체 만큼의 행
print(test_array)
```

```python
train_input = train_input.reshape(-1, 1)  # 1차원 배열 -> 열이 1개인 2차원 배열로 변환
test_input = test_input.reshape(-1, 1)  # 1차원 배열 -> 열이 1개인 2차원 배열로 변환
print(train_input.shape, test_input.shape)
```

```python
# K-최근접 이웃 회귀로 학습 하기
from sklearn.neighbors import KNeighborsRegressor  # KNeighborsClassifier - 분류, KNeighborsRegressor - 회귀

knr = KNeighborsRegressor()

# 훈련
knr.fit(train_input, train_target)  # 학습 데이터 - 농어 길이, 정답 데이터(타겟) - 농어 무게
```

```python
# 결정 계수
print(knr.score(train_input, train_target))
print(knr.score(test_input, test_target))
```

```python
knr.predict([[41]]) # 길이가 41cm 농어의 무게는?
```

```python
# 예측 모델의 오차 - 평균 절대값 오차(MAE), 평균 제곱 오차(MSE)
from sklearn.metrics import mean_absolute_error

test_prediction = knr.predict(test_input)  # 테스트 세트의 길이로 무게를 예측 
print(test_prediction)
print(test_target)

mae = mean_absolute_error(test_target, test_prediction)
print(mae)
```

```python
# 과소 적합인 경우 - 과소 적합 : 테스트 세트의 결정 계수가 훈련 세트 결정 계수보다 큰 경우 
# 전반적으로 결정계수가 둘다 작은 경우
# 모델을 좀더 복잡하게 만들면 해결 된다!
# K-최근접 이웃 알고리즘에서는 이웃의 갯수를 줄이면 모델이 복잡해 진다!
knr.n_neighbors = 3  # 기존 5 -> 3 이웃 개수를 변경, 모델이 복잡해지는 것
knr.fit(train_input, train_target)
```

```python
print(knr.score(train_input, train_target)) # 훈련 세트의 결정 계수
print(knr.score(test_input, test_target)) # 테스트 세트의 결정 계수
```
